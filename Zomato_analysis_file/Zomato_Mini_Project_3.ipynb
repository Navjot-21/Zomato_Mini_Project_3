```python
cuisine_rating = df[['Cuisines', 'Aggregate rating']].copy()

cuisine_rating['Cuisines'] = cuisine_rating['Cuisines'].str.split(',')

cuisine_rating = cuisine_rating.explode('Cuisines')

cuisine_rating['Cuisines'] = cuisine_rating['Cuisines'].str.strip()

cuisine_rating.head()
```

```python
highest_rated = (
    cuisine_rating
    .groupby('Cuisines')['Aggregate rating']
    .mean()
    .sort_values(ascending=False)
)

highest_rated.head(10)
```

Some cuisines may appear only once and have a perfect rating of 5.0, which can be misleading. Filter out cuisines with very few restaurants.

```python
cuisine_stats = cuisine_rating.groupby('Cuisines').agg(
    Average_Rating=('Aggregate rating','mean'),
    Restaurant_Count=('Aggregate rating','count')
)

# Keep cuisines served by at least 20 restaurants
highest_rated = (
    cuisine_stats[cuisine_stats['Restaurant_Count'] >= 20]
    .sort_values('Average_Rating', ascending=False)
    .head(10)
)

highest_rated
```

# Calculate Skewness

```python
skewness = df['Aggregate rating'].skew()

print("Skewness =", round(skewness,3))
```

### Interpretation of Skewness

- Skewness > 0 : Positively (Right) Skewed
- Skewness < 0 : Negatively (Left) Skewed
- Skewness ≈ 0 : Approximately Symmetric

Skewness helps understand whether restaurant ratings are concentrated towards higher or lower values.

The skewness value of -0.954 indicates that the distribution of restaurant ratings is negatively skewed. This is primarily due to the large number of restaurants with an aggregate rating of 0, representing restaurants that have not yet received customer ratings. Apart from these unrated restaurants, most rated restaurants are concentrated between ratings of 3.0 and 4.2, indicating generally positive customer feedback.

# Detect Outliers using IQR

```python
Q1 = df['Aggregate rating'].quantile(0.25)
Q3 = df['Aggregate rating'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Q1 :", Q1)
print("Q3 :", Q3)
print("IQR :", IQR)
print("Lower Limit :", lower_limit)
print("Upper Limit :", upper_limit)
```

```python
outliers = df[
    (df['Aggregate rating'] < lower_limit) |
    (df['Aggregate rating'] > upper_limit)
]

print("Number of Outliers :", len(outliers))
```

The IQR method identified 2,148 observations as outliers. Most of these correspond to restaurants with an aggregate rating of 0, which indicates that they are "Not Rated" rather than poorly rated. Therefore, these outliers represent missing customer feedback instead of abnormal restaurant performance.

```python
outliers.head()
```

### Observation

The Interquartile Range (IQR) method identifies ratings that lie significantly below or above the normal range.

These observations may represent unusually low-rated or exceptionally high-rated restaurants.

### Rating Analysis (Rated Restaurants Only)

The original dataset contains many restaurants with an aggregate rating of **0**, which indicates that they have **not yet been rated** by customers. Including these restaurants can distort the distribution and statistical analysis.

Therefore, to better understand actual customer feedback, only restaurants with an aggregate rating greater than zero are considered in this analysis.

```python
# Keep only restaurants that have received ratings
rated_df = df[df['Aggregate rating'] > 0]

plt.figure(figsize=(10,6))

sns.histplot(
    data=rated_df,
    x='Aggregate rating',
    bins=20,
    color='royalblue',
    edgecolor='black'
)

plt.title("Distribution of Ratings (Rated Restaurants Only)")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
skewness = rated_df['Aggregate rating'].skew()

print("Skewness =", round(skewness,3))

if skewness > 0:
    print("The distribution is Positively (Right) Skewed.")
elif skewness < 0:
    print("The distribution is Negatively (Left) Skewed.")
else:
    print("The distribution is Approximately Symmetric.")
```

```python
Q1 = rated_df['Aggregate rating'].quantile(0.25)
Q3 = rated_df['Aggregate rating'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = rated_df[
    (rated_df['Aggregate rating'] < lower_limit) |
    (rated_df['Aggregate rating'] > upper_limit)
]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Limit:", round(lower_limit,2))
print("Upper Limit:", round(upper_limit,2))
print("Number of Outliers:", len(outliers))

plt.figure(figsize=(10,3))

sns.boxplot(
    x=rated_df['Aggregate rating'],
    color='lightgreen'
)

plt.title("Boxplot of Rated Restaurant Ratings")

plt.show()
```

# Task 10: Feature Engineering

## Objective

Feature engineering is the process of creating new variables from existing data to improve analysis and interpretation.

In this task, two new features are created:

- **Cost Category** based on the average cost for two people.
- **Rating Category** based on the aggregate customer rating.

These engineered features simplify the analysis by grouping restaurants into meaningful categories.

```python
# Create Cost Category
rated_df = rated_df.copy()
def cost_category(cost):
    if cost <= 500:
        return 'Budget'
    elif cost <= 1500:
        return 'Moderate'
    elif cost <= 3000:
        return 'Premium'
    else:
        return 'Luxury'

rated_df['Cost Category'] = rated_df['Average Cost for two'].apply(cost_category)

rated_df[['Average Cost for two', 'Cost Category']].head()
```

### Cost Category

Restaurants are classified into four pricing categories:

- **Budget:** ₹0 – ₹500
- **Moderate:** ₹501 – ₹1500
- **Premium:** ₹1501 – ₹3000
- **Luxury:** Above ₹3000

These categories make it easier to compare restaurants based on pricing.

```python
# Create Rating Category

def rating_category(rating):
    if rating < 2.5:
        return 'Poor'
    elif rating < 3.5:
        return 'Average'
    elif rating < 4.5:
        return 'Good'
    else:
        return 'Excellent'

rated_df['Rating Category'] = rated_df['Aggregate rating'].apply(rating_category)

rated_df[['Aggregate rating', 'Rating Category']].head()
```

### Rating Category

Restaurants are grouped according to their customer ratings:

- **Poor:** Rating below 2.5
- **Average:** 2.5 – 3.49
- **Good:** 3.5 – 4.49
- **Excellent:** 4.5 and above

This categorization simplifies the interpretation of restaurant performance.

```python
# Analyze Cost Categories

cost_counts = rated_df['Cost Category'].value_counts()

cost_counts
```

Restaurants without online delivery have an average rating of **3.47**, while restaurants providing online delivery have an average rating of **3.38**. The small difference suggests that online delivery alone does not significantly affect customer ratings.

### 5. Most rated restaurants fall into the Good category.

The majority of rated restaurants have customer ratings between **3.5 and 4.5**, indicating that most restaurants on the platform provide satisfactory dining experiences. Very few restaurants are classified as Poor or Excellent.
